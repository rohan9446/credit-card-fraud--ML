```python
import numpy as np
import pandas as pd
import re
import matplotlib.pyplot as plt
from matplotlib import style
style.use('ggplot')
import seaborn as sns



from sklearn.model_selection import train_test_split, GridSearchCV, RandomizedSearchCV
from imblearn.ensemble import EasyEnsembleClassifier,  RUSBoostClassifier, BalancedBaggingClassifier, BalancedRandomForestClassifier

from sklearn.metrics import accuracy_score, classification_report, confusion_matrix, ConfusionMatrixDisplay

import warnings
warnings.filterwarnings('ignore')
```

#Loading Data

```python
df = pd.read_csv('creditcard.csv')
df.head()
```

```python
X=df.drop(['Class'],axis=1).values
y=df['Class'].values
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=42)
```

#Essemble classifier

```python
easy=EasyEnsembleClassifier(random_state=42)
easy.fit(X_train,y_train)
```

```python
y_pred = easy.predict(X_test)
y_pred2=easy.predict(X_train)
```

```python
print("train_accuracy:")
print(accuracy_score(y_pred2,y_train)*100)
print('----------------------------------')
print("test_accuracy:")
print(accuracy_score(y_pred,y_test)*100)
print('----------------------------------')
print(classification_report(y_test, y_pred))
print('----------------------------------')
plt.style.use('fivethirtyeight')
cm = confusion_matrix(y_test, y_pred, labels=easy.classes_)
disp = ConfusionMatrixDisplay(confusion_matrix = cm, display_labels=easy.classes_)
disp.plot(cmap='Blues', values_format='d')
plt.title('EasyEnsembleClassifier',fontsize=16, fontweight='bold',pad=20)
plt.show()
```

```python
clf = RUSBoostClassifier(random_state=0)
clf.fit(X_train,y_train)
```

```python
y_pred = clf.predict(X_test)
y_pred2=clf.predict(X_train)
```

```python
print("train_accuracy:")
print(accuracy_score(y_pred2,y_train)*100)
print('----------------------------------')
print("test_accuracy:")
print(accuracy_score(y_pred,y_test)*100)
print('----------------------------------')
print(classification_report(y_test, y_pred))
print('----------------------------------')
plt.style.use('fivethirtyeight')
cm = confusion_matrix(y_test, y_pred, labels=clf.classes_)
disp = ConfusionMatrixDisplay(confusion_matrix = cm, display_labels=clf.classes_)
disp.plot(cmap='Blues', values_format='d')
plt.title('RUSBoostClassifierr',fontsize=16, fontweight='bold',pad=20)
plt.show()
```

```python
bbc = BalancedBaggingClassifier(random_state=42)
bbc.fit(X_train, y_train)
```

```python
y_pred = bbc.predict(X_test)
y_pred2=bbc.predict(X_train)
```

```python
print("train_accuracy:")
print(accuracy_score(y_pred2,y_train)*100)
print('----------------------------------')
print("test_accuracy:")
print(accuracy_score(y_pred,y_test)*100)
print('----------------------------------')
print(classification_report(y_test, y_pred))
print('----------------------------------')
plt.style.use('fivethirtyeight')
cm = confusion_matrix(y_test, y_pred, labels=bbc.classes_)
disp = ConfusionMatrixDisplay(confusion_matrix = cm, display_labels=bbc.classes_)
disp.plot(cmap='Blues', values_format='d')
plt.title('BalancedBaggingClassifier',fontsize=16, fontweight='bold',pad=20)
plt.show()
```

```python
clf2 = BalancedRandomForestClassifier(
    sampling_strategy="all", replacement=True, max_depth=2, random_state=0)
clf2.fit(X_train, y_train)
```

```python
y_pred = clf2.predict(X_test)
y_pred2=clf2.predict(X_train)
```

```python
print("train_accuracy:")
print(accuracy_score(y_pred2,y_train)*100)
print('----------------------------------')
print("test_accuracy:")
print(accuracy_score(y_pred,y_test)*100)
print('----------------------------------')
print(classification_report(y_test, y_pred))
print('----------------------------------')
plt.style.use('fivethirtyeight')
cm = confusion_matrix(y_test, y_pred, labels=clf2.classes_)
disp = ConfusionMatrixDisplay(confusion_matrix = cm, display_labels=clf2.classes_)
disp.plot(cmap='Blues', values_format='d')
plt.title('BalancedRandomForestClassifier',fontsize=16, fontweight='bold',pad=20)
plt.show()
```
